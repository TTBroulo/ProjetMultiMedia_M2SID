```python
df = pd.read_csv('/content/gdrive/MyDrive/Projet_Multimedia/download/CSV/indice_vid.csv')
files = df['name']
```

```python
videos = ['video'+ x+'.mp4' for x in files]
```

```python
dict_videos = {}
mod = 10
for count, video in tqdm(enumerate(videos)) :
  if count % 20 == mod :
    result_yolo = !python detect.py --weights yolov7.pt --save-txt --conf 0.25 --img-size 640 --source /content/gdrive/MyDrive/Projet_Multimedia/videos/{video}
    str_dict = result_yolo[-1]
    str_dict = re.sub("\'", "\"", str_dict)
    dict = json.loads(str_dict)
    dict_videos[video] = dict

str_dict_all = re.sub("\'", "\"", str(dict_videos))
# dict_videos_json = json.loads(str_dict)

# Spécifier le chemin du fichier JSON

chemin_fichier = "/content/gdrive/MyDrive/Projet_Multimedia/download/yolo_json/objets_videos_mod_"+str(mod)+".json"

# Ouvrir le fichier en mode écriture et écrire les données JSON
with open(chemin_fichier, 'w') as fichier_json:
    fichier_json.write(str_dict_all)
```
